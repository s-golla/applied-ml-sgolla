# Titanic Histogram Exploration Notebook

This notebook explores the Titanic dataset using histograms to visualize key variables.

## Step 1: Import Required Libraries

We'll import the necessary libraries for data analysis and visualization:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load the Dataset

Load the Titanic dataset from the CSV file:

```python
# Make sure 'data\train.csv' is downloaded from https://www.kaggle.com/competitions/titanic/data
df = pd.read_csv(r'C:\\Repos\\applied-ml-sgolla\\data\\train.csv')
```

## Step 3: Visualize Target Variable

Create a histogram to visualize the distribution of the target variable 'Survived':

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Histogram of Target Variable: Survived')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Step 4: Visualize Age Distribution

Create a histogram to visualize the distribution of passenger ages:

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Histogram of Input Variable: Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Step 5: Missing Values Analysis

Check for missing values in the dataset:

```python
missing_summary = df.isnull().sum()
print("Missing Values Summary:\n", missing_summary)
```
